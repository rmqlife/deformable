```python
import rosbag
import cv2
import os, re
from sensor_msgs.msg import Image
from cv_bridge import CvBridge
import numpy as np
# conda install libgcc

t1 = '/yumi/ikSloverVel_controller/state'
t2 = '/yumi/ikSloverVel_controller/ee_cart_position'
t3 = '/camera/image/rgb_611205001943'
t4 = '/camera/image/depth_611205001943'

bag_name_list = ['bags_2/towel_1.bag','bags_2/towel_2.bag','bags_2/towel_3.bag','bags_2/towel_4.bag']
bag_name_list = ['bags_3/test.bag']

bridge = CvBridge()
topic_pos = '/yumi/ikSloverVel_controller/state'
topic_vel = '/yumi/ikSloverVel_controller/ee_cart_position'
topic_rgb = '/camera/image/rgb_611205001943'
topic_depth = '/camera/image/depth_611205001943'

count = 0
tt_pos = np.array([])
tt_vel = np.array([])

os.chdir('/home/jiabiao/work/deformable/')

bag_folder = 'bags_3/'
# os.mkdir(bag_folder+'rgb')
# os.mkdir(bag_folder+'depth')

for bag_name in bag_name_list:
    bag = rosbag.Bag(bag_name)
    print bag
    for topic, msg, t in bag.read_messages(topics=[topic_pos,topic_vel,topic_depth, topic_rgb]):
        if topic == topic_rgb:
            count = count + 1
            cv_img = bridge.imgmsg_to_cv2(msg)
            cv_img = cv2.cvtColor(cv_img,cv2.COLOR_BGR2RGB)
            cv2.imwrite(os.path.join("bags_3/rgb/%04i.jpg" % count), cv_img)
        
        if topic == topic_depth:
            cv_img = bridge.imgmsg_to_cv2(msg)
            cv2.imwrite(os.path.join("bags_3/depth/%04i.png" % count), cv_img)
        
        if topic == topic_pos:
            pos = np.array(msg.data)
            tt_pos = np.vstack((tt_pos, pos)) if tt_pos.size else pos
            
        if topic == topic_vel:
            vel = np.array(msg.data)
            tt_vel = np.vstack((tt_vel, vel)) if tt_vel.size else vel
            
    print tt_pos.shape, tt_vel.shape
    
pack_name = "bags_3/data.npz"
np.savez(pack_name, vel = tt_vel, pos = tt_pos)
```

```python
from keras.layers import Dense, Activation, Dropout, MaxPooling1D
from keras.models import Sequential
import keras

X= feat
y = np.atleast_2d(pos[:,1]).T

sep = 1000
X_train = X[0:sep,:]
X_test = X[sep:-1,:]
y_train = y[0:sep,:]
y_test = y[sep:-1,:]

# from sklearn import decomposition
# pca = decomposition.PCA(500)
# pca.fit(X_train)
# X_train = pca.transform(X_train)
# X_test =pca.transform(X_test)
print X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# # http://machinelearningmastery.com/regression-tutorial-keras-deep-learning-library-python/
model = Sequential()
model.add(Dense(X_train.shape[0], input_dim=X_train.shape[1], kernel_initializer='normal', activation='relu'))
model.add(Dense(10, kernel_initializer='normal', activation='relu'))

model.add(Dense(y.shape[1], kernel_initializer='normal'))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train,y_train,epochs=20, batch_size=500, verbose=2, shuffle=True, validation_data=(X_test,y_test))

y_pred = model.predict(X_test)
# print y_test.shape
for i in range(y_test.shape[0]):
    print i
    print y_pred[i]
    print y_test[i]
```
