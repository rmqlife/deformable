```python
from random import gauss
import numpy as np

def rand_unit_vector(dims,length = 1):
    vec = [gauss(0, 1) for i in range(dims)]
    mag = sum(x**2 for x in vec) ** .5
    ret = [x/mag for x in vec]
    ret = np.array(ret)*length
    return ret

def rand_sphere(center, radius=0.3):
    vec = rand_unit_vector(3,radius)
    return center+vec

def in_sphere(pos,center,radius):
    dist = sum((pos-center)**2)**0.5
    if dist<=radius:
        return True
    else:
        return False

center = np.array([0.3,0.2,0.1])
for i in range(200):
    pos = rand_sphere(center,0.3)
    print sum((pos-center)**2)**0.5,pos, in_sphere(pos,center,0.301)
```
