```python
import os, cv2
import matplotlib.pyplot as plt
import numpy as np
# draw bounding box
def draw_bbox(im,bbox):
    rp1 = (int(bbox[0]), int(bbox[1]))
    rp2 = (int(bbox[0]+bbox[2]), int(bbox[1]+bbox[3]))
    print rp1, rp2
    return cv2.rectangle(im.copy(), rp1, rp2, (255,0,0),2)

def show_bbox(im,bbox):
    plt.imshow(draw_bbox(im,bbox))
    plt.show()
    return

os.chdir('/home/jiabiao/work/deformable/tracking/rgb/')
filelist = []
for dirpath, dirnames, filenames in os.walk('.'):
    for filename in filenames:
        fn = os.path.join(dirpath,filename)
        filelist.append(fn)
filelist = sorted(filelist)

# initialize the frame
im = cv2.imread(filelist[0],1)
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
# Set up tracker.
# Instead of MIL, you can also use BOOSTING, KCF, TLD, MEDIANFLOW or GOTURN
tracker = cv2.Tracker_create("BOOSTING")
# Define an initial bounding box
init_bbox = (410, 180, 30, 30)
# Show tracker
show_bbox(im,init_bbox)
# Initialize tracker with first frame and bounding box
ret = tracker.init(im, init_bbox)

tt_bbox = np.array([])
for i in range(0,len(filelist),5):
    im = cv2.imread(filelist[i],1)
    im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
    ret, bbox = tracker.update(im)
    bbox = np.array(bbox)
    tt_bbox = np.vstack((tt_bbox,bbox)) if tt_bbox.size else bbox
    print filelist[i], tt_bbox.shape, bbox
    show_bbox(im,bbox)
```
